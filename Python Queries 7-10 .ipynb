```python
import sqlite3
import os
import pandas as pd
```

```python
##First connect to the database

first_db = sqlite3.connect("owner_prod.db")
```

```python
## Now let's create a cursor

curs = first_db.cursor()
```

```python
curs.execute('''DROP TABLE IF EXISTS owner_products''')
```

```python
##good practice to include a drop table if
curs.execute('''DROP TABLE IF EXISTS owner_products''')

##Create the table

curs.execute('''CREATE TABLE owner_products(
            owner INTEGER,
            upc INTEGER,
            description TEXT,
            dept_name TEXT,
            spend NUMERIC,
            items INTEGER,
            trans INTEGER)''')
```

```python
#read in this file, and create a list of lists, append these lists to owner_prod list.
owner_prod = []
with open("owner_products.txt",'r') as infile :
    next(infile)
    
    for line in infile :
        line = line.strip().split("\t")
        owner_prod.append(line)
        
        ##Now let's add owner prod to our table
        curs.execute('''
            INSERT INTO owner_products(owner,upc,description,dept_name,
            spend,items,trans)VALUES(?,?,?,?,?,?,?)''',line)

first_db.commit()
```

```python
##Now let's run the final query 
curs.execute('''SELECT description,
                    dept_name,
                    sum(spend) as total_spend
                    FROM owner_products
                    WHERE dept_name = "REF GROCERY"
                    OR dept_name = "PACKAGED GROCERY"
                    GROUP BY description
                    ORDER BY total_spend DESC
                    LIMIT 10''')



```

```python
#Lets get our results from the above query in a list
output = curs.fetchall()
```
